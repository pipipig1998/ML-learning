# 第三章线性回归



## 模型

### 训练数据
&ensp;&ensp;房屋面积$x_1$,房龄为$x_2$，售价为$y$,建立$x_1,x_2$的一个映射
$$
\hat y=x_1w_1+x_2w_2+b
$$<br>
$w_1$和$w_2$是权重，$b$是偏差，且都为均量
### 损失函数
$$
    l^{i}(w_1,w_2,b)=\frac{1}{2}(\hat y^{i}-y^{i})^2\\
    l(w_1,w_2.b)=\frac{1}{n}\Sigma_{i=1}^{n}l^{(i)}(w_1,w_2,b)=\frac{1}{n}\Sigma_{i=1}^{n}\frac{1}{2}(x_1^{(i)} w_1+x_2^{(i)} w_2 +b-y^{(i)})^2
$$

我们希望找到语族模型参数，使得其满足$argmin_{w_1,w_2,b}l(w_1,w_2,b)$

### 优化算法
&ensp;&ensp;
当模型和损失函数形式角为简单，可以把上边误差最小化问题的解可以直接用公式表达出来，这种解成为**解析解**。然而大多数深度学习的模型并没有解析解，只能通过优化算法有限次迭代模型参数来尽可能的降低损失函数的值，这种解叫做**数值解**。<br>
&ensp;&ensp;
在求数值解的优化算法中，小批量随机梯度下降，算法很简单：<br>
1. 先短曲一组模型参数的初始值，如随机选取
2. 对参数进行多次迭代，每次迭代都可能降低损失函数的值。
3. 每次迭代，先随机均匀采样一个固定数目的样本组成的小批量的$\beta$,然后求小批量样本中模型参数 

$$
    w_1=w_1-\frac{\eta}{|\beta|}\Sigma_{i\in \beta}x_1^{i}(x_1^{i}w_1+x_2^{i}w_2+b-y_{i})\\
    w_2=w_2-\frac{\eta}{|\beta|}\Sigma_{i\in \beta}x_2^{i}(x_1^{i}w_1+x_2^{i}w_2+b-y_{i})\\
    b=b-\frac{\eta}{|\beta|}\Sigma_{i\in\beta}(x_1^{i}w_1+x_2^{i}w_2+b-y_{i})
$$

**值得注意一点，这里学习率$\eta$和$\beta$都是人为设定的，并不能通过学习习得因此叫做超参数<br>调参指得就是调节超参数**

```python
# 先考虑以下两个向量相加的两种方法
from mxnet import nd
from time import time

a=nd.ones(shape=1000)
b=nd.ones(shape=1000)

start=time()
c=nd.zeros(shape=1000)
for i in range(1000):
    c[i]=a[i]+b[i]
end=time()
print(end-start)
# 0.4323000907897949

start=time()
c=a+b
print(c)
end=time()
print(end-start)
# 0.0009975433349609375
```

很明显，矢量运算更快捷，将预测价格变为矢量来进行运算
重写损失函数,设模型输出$\hat y =R^{n\times 1}$,样本特征为$\chi\in R^{n\times d}$,权重$w\in R^{d\times 1}$,偏差$b\in R$。批量数据标签$y\in R^{n\times 1}$,设模型参数为$\theta=[w_1,w_2,b]^{2}$，可以重写损失函数:
$$
l(\theta)=\frac{1}{2n}(\hat y-y)^T(\hat y - y)\\
\theta=\theta-\frac{\eta}{|\beta|}\Sigma_{i\in\beta}\nabla_{\theta}l^{(i)}(\theta)
$$

**开始正式的线性模型**

## 代码
### 导包

```python
from matplotlib import pyplot as plt
from mxnet import autograd,nd
from IPython import display
import random
```

### 生成数据集
设训练集样本为$1000$，特征数为$2$，随机生成样本$\chi \in R^{1000\times 2}$,真实权重为$w=[2,-3.4]^T$和偏差$b=4.2$以及一个随机噪声$\varepsilon$来生成标签

```python
num_input=2
num_example = 1000
true_w = [2.,-3.14]
true_b = 4.2
# 建立特征的正态分布,方差为1，高1000宽2
features = nd.random.normal(scale=1,shape=(num_example,num_input))
# 生成标签
labels = true_w[0]*feature[:,0]+true_w[1]*feature[:,1]+true_b
# 加入噪声
labels +=nd.random.normal(scale=0.01,shape=labels.shape)
# print(labels.shape)
```

```python
#画图查看
plt.figure(figsize=(3.5,3.5))
# 记得转换成numpy数组
plt.scatter(features[:,1].asnumpy(),labels.asnumpy())
plt.show()
```

```python
# 实验梯度下降函数的书写
batch_size=10
num=0
for X,y in data_iter(batch_size,features,labels):
    print(X,y)
```

### 初始化参数模型

```python
#创建权重矩阵W和偏执b
w=nd.random.normal(scale=0.01,shape=(num_input,1))
b=nd.zeros(shape=(1,))
#利用自动求导机制来进行随机梯度下降
w.attach_grad()
b.attach_grad()

```

### 定义损失函数

```python
#定义损失函数
def squared_loss(y_hat,y):
    return (y_hat-y)**2/2
```

```python
# 定义梯度下降
#[w,b],学习率，长度
def sgd(params,lr,batch_size):
    for param in params:
            param[:]=param-lr*param.grad/batch_size
#y_hat
def linreg(X,w,b):
    return nd.dot(X,w)+b
```

### 开始训练

```python
#学习率
lr =0.03
# 迭代周期
num_epochs=3
# 全连接层
net=linreg
# 损失函数
loss=squared_loss
for epochs in range(num_epochs):
    for X,y in data_iter(batch_size,features,labels):
        with autograd.record():
            l=loss(net(X,w,b),y)
        l.backward()
        sgd([w,b],lr,batch_size)
    train_l=loss(net(features,w,b),labels)
    print('epoch %d ,loss %f'%(epochs+1,train_l.mean().asnumpy()))
```
